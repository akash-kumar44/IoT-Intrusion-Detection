```python
# â”€â”€ Cell 1: Imports â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 50)
pd.set_option('display.float_format', '{:.4f}'.format)

print("Libraries loaded")
```

```python
# â”€â”€ Cell 2: Column Names (from NUSWNB15_features.csv) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
FEATURE_NAMES = [
    'srcip', 'sport', 'dstip', 'dsport', 'proto', 'state', 'dur',
    'sbytes', 'dbytes', 'sttl', 'dttl', 'sloss', 'dloss', 'service',
    'Sload', 'Dload', 'Spkts', 'Dpkts', 'swin', 'dwin', 'stcpb', 'dtcpb',
    'smeansz', 'dmeansz', 'trans_depth', 'res_bdy_len', 'Sjit', 'Djit',
    'Stime', 'Ltime', 'Sintpkt', 'Dintpkt', 'tcprtt', 'synack', 'ackdat',
    'is_sm_ips_ports', 'ct_state_ttl', 'ct_flw_http_mthd', 'is_ftp_login',
    'ct_ftp_cmd', 'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ltm',
    'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm',
    'attack_cat', 'label'
]

print(f"Total columns defined: {len(FEATURE_NAMES)}")
```

```python
# â”€â”€ Cell 3: Load & Combine â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
from google.colab import drive
drive.mount('/content/drive')

DATA_PATH = "/content/drive/My Drive/IoT Intrusion Detection/data/"


dfs = []
for i in range(1, 5):
    filepath = f'{DATA_PATH}UNSW-NB15_{i}.csv'
    temp = pd.read_csv(
        filepath,
        header=None,          # raw files have no header row
        names=FEATURE_NAMES,
        low_memory=False,
        encoding='cp1252'
    )
    print(f"  UNSW-NB15_{i}.csv -> {temp.shape[0]:,} rows")
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True)
print(f"\nCombined shape: {df.shape[0]:,} rows Ã— {df.shape[1]} columns")
```

```python
# â”€â”€ Cell 4: First Look â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("=== Shape ===")
print(df.shape)

print("\n=== First 3 rows ===")
display(df.head(3))

print("\n=== Data types ===")
print(df.dtypes)

print("\n=== Label distribution ===")
label_counts = df['label'].value_counts()
print(label_counts)
print(f"\nAttack rate: {label_counts[1] / len(df) * 100:.1f}%")
```

```python
# â”€â”€ Cell 5: Null Check â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
null_counts = df.isnull().sum()
null_pct = (null_counts / len(df) * 100).round(2)

null_summary = pd.DataFrame({
    'null_count': null_counts,
    'null_pct': null_pct
}).query('null_count > 0').sort_values('null_pct', ascending=False)

if null_summary.empty:
    print("No null values found âœ“")
else:
    print(f"{len(null_summary)} columns have nulls:\n")
    display(null_summary)
```

```python
# â”€â”€ Cell 6: Save Combined Raw Data â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df.to_csv('/content/drive/My Drive/IoT Intrusion Detection/data/UNSW_NB15_combined_raw.csv', index=False)
print("Saved -> /content/drive/My Drive/IoT Intrusion Detection/data/UNSW_NB15_combined_raw.csv")
```
